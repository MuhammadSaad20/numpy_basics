**Looping NumPy**

```python
k=np.arange(12).reshape(3,4)
k
```

```python
for row in k:
    for cell in row:
        print(cell)
```

```python
for cell in k.flatten():
    print(cell)
```

```python
for x in np.nditer(k,order='C'):
    print(x)#print row wise
```

```python
for x in np.nditer(k,order='F'):
    print(x)#print colum wise
```

```python
for x in np.nditer(k,order='F',flags=['external_loop']):
    print(x)#Print colum wise
```

```python
for x in np.nditer(k,op_flags=['readwrite']):
    x[...]=x*x
```
